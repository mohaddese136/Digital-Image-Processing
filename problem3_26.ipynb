create checkboard image

```python
import numpy as np
import cv2

# Set the size of the checkerboard pattern
width, height = 400, 400
num_squares = 8  # Number of squares per row and column
square_size = width // num_squares

# Create an empty white image
checkerboard = np.ones((height, width, 3), dtype=np.uint8) * 255

# Fill the squares with black or white color to create the checkerboard pattern
for i in range(num_squares):
    for j in range(num_squares):
        if (i + j) % 2 == 1:
            checkerboard[i * square_size:(i + 1) * square_size, j * square_size:(j + 1) * square_size] = 0

# Save or display the checkerboard image
cv2.imwrite('checkerboard.png', checkerboard)




#create fisrt pic

# Create an empty image
half_black_white = np.zeros((height, width, 3), dtype=np.uint8)

# Fill the left half with white and the right half with black
half_black_white[:, :width // 2] = [255, 255, 255]  # White (BGR color)
half_black_white[:, width // 2:] = [0, 0, 0]  # Black (BGR color)

# Save or display the half black and half white image
cv2.imwrite('half_black_white.png', half_black_white)
```

plot histograms

```python
import matplotlib.pyplot as plt

image = cv2.imread('half_black_white.png')

# Define a 3x3 blur kernel
kernel = np.ones((3, 3), np.float32) / 9

# Apply the convolution operation to blur the image
blurred_image = cv2.filter2D(image, -1, kernel)

# Compute the histogram of the blurred image
hist = cv2.calcHist([blurred_image], [0], None, [256], [0, 256])

# Plot the histogram
plt.figure(figsize=(8, 6))
plt.title('Histogram of Blurred Image')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.plot(hist)
plt.xlim([0, 256])
plt.grid()

# Save the blurred image
cv2.imwrite('output_blurred_image.jpg', blurred_image)

# Show the histogram plot
plt.show()
```
